```python
import ffmpeg
import os
import tempfile
from pathlib import Path

def get_frame_count_ffmpeg_python(video_path):
    """
    Get frame count using ffmpeg-python.
    """
    if not os.path.exists(video_path):
        raise FileNotFoundError(f"Video file not found: {video_path}")
    
    try:
        probe = ffmpeg.probe(video_path, v='quiet', select_streams='v:0')
        video_stream = next((stream for stream in probe['streams'] if stream['codec_type'] == 'video'), None)
        
        if video_stream is None:
            raise ValueError("No video stream found")
        
        # Try to get frame count from metadata
        if 'nb_frames' in video_stream:
            return int(video_stream['nb_frames'])
        
        # If nb_frames not available, calculate from duration and fps
        duration = float(video_stream['duration'])
        fps_str = video_stream['r_frame_rate']
        fps_num, fps_den = map(int, fps_str.split('/'))
        fps = fps_num / fps_den
        
        return int(duration * fps)
        
    except Exception as e:
        raise ValueError(f"Could not process video file: {e}")

def scale_video_frames_precise(input_path, output_path, target_frames, maintain_fps=True):
    """
    More precise method using frame extraction and reconstruction.
    Maintains original FPS while changing duration proportionally.
    
    Args:
        input_path (str): Path to input MP4 file
        output_path (str): Path for output MP4 file
        target_frames (int): Desired number of frames
        maintain_fps (bool): If True, keeps original FPS and changes duration
    
    Returns:
        dict: Information about the scaling operation
    """
    if not os.path.exists(input_path):
        raise FileNotFoundError(f"Input video file not found: {input_path}")
    
    current_frames = get_frame_count_ffmpeg_python(input_path)
    scale_factor = target_frames / current_frames
    
    print(f"Original frames: {current_frames}")
    print(f"Target frames: {target_frames}")
    print(f"Scale factor: {scale_factor:.3f}")
    
    # Get video properties
    probe = ffmpeg.probe(input_path, v='quiet')
    video_stream = next((stream for stream in probe['streams'] if stream['codec_type'] == 'video'), None)
    
    # Get original FPS
    fps_str = video_stream['r_frame_rate']
    fps_num, fps_den = map(int, fps_str.split('/'))
    original_fps = fps_num / fps_den
    
    print(f"Original FPS: {original_fps}")
    
    # Calculate new duration based on target frames and original FPS
    new_duration = target_frames / original_fps
    original_duration = float(video_stream['duration'])
    
    print(f"Original duration: {original_duration:.3f}s")
    print(f"New duration: {new_duration:.3f}s")
    
    with tempfile.TemporaryDirectory() as temp_dir:
        # Extract all frames from original video
        frame_pattern = os.path.join(temp_dir, 'frame_%06d.png')
        (
            ffmpeg
            .input(input_path)
            .output(frame_pattern, vcodec='png')
            .overwrite_output()
            .run(quiet=True)
        )
        
        # Create frame list for target number of frames
        frame_list_path = os.path.join(temp_dir, 'frame_list.txt')
        frame_duration = 1.0 / original_fps  # Duration per frame at original FPS
        
        with open(frame_list_path, 'w') as f:
            for i in range(target_frames):
                # Map target frame index to source frame index
                source_frame_idx = int((i / target_frames) * current_frames) + 1
                source_frame_idx = min(source_frame_idx, current_frames)  # Clamp to max
                frame_file = f'frame_{source_frame_idx:06d}.png'
                f.write(f"file '{frame_file}'\n")
                f.write(f"duration {frame_duration}\n")
            
            # Add the last frame without duration to properly end the video
            f.write(f"file 'frame_{current_frames:06d}.png'\n")
        
        # Reconstruct video from frame list with original FPS
        temp_output = os.path.join(temp_dir, 'temp_video.mp4')
        (
            ffmpeg
            .input(frame_list_path, format='concat', safe=0)
            .output(
                temp_output, 
                vcodec='libx264', 
                pix_fmt='yuv420p',
                r=original_fps,  # Explicitly set FPS to original
                vsync='cfr'      # Constant frame rate
            )
            .overwrite_output()
            .run(quiet=True)
        )
        
        # Handle audio if present
        audio_streams = [stream for stream in probe['streams'] if stream['codec_type'] == 'audio']
        
        if audio_streams:
            print("Processing audio track...")
            # Get audio and stretch/compress it to match new video duration
            audio_tempo = original_duration / new_duration
            
            # FFmpeg atempo filter has limits (0.5 to 2.0), so we might need to chain multiple filters
            if audio_tempo > 2.0:
                # Need to slow down audio significantly - chain multiple atempo filters
                tempo_filters = []
                remaining_tempo = audio_tempo
                while remaining_tempo > 2.0:
                    tempo_filters.append('atempo=0.5')
                    remaining_tempo *= 0.5
                tempo_filters.append(f'atempo={1/remaining_tempo}')
                audio_filter = ','.join(tempo_filters)
            elif audio_tempo < 0.5:
                # Need to speed up audio significantly - chain multiple atempo filters
                tempo_filters = []
                remaining_tempo = audio_tempo
                while remaining_tempo < 0.5:
                    tempo_filters.append('atempo=2.0')
                    remaining_tempo *= 2.0
                tempo_filters.append(f'atempo={remaining_tempo}')
                audio_filter = ','.join(tempo_filters)
            else:
                # Single atempo filter is sufficient
                audio_filter = f'atempo={audio_tempo}'
            
            print(f"Audio tempo adjustment: {audio_tempo:.3f} (filter: {audio_filter})")
            
            # Combine video with adjusted audio
            video_input = ffmpeg.input(temp_output)
            audio_input = ffmpeg.input(input_path)
            
            (
                ffmpeg
                .output(
                    video_input['v'],
                    audio_input['a'].filter('aformat', 'sample_fmts=fltp:sample_rates=48000').filter('aresample', 48000).filter(audio_filter.split(',')[0] if ',' in audio_filter else audio_filter),
                    output_path,
                    vcodec='copy',
                    acodec='aac',
                    audio_bitrate='128k'
                )
                .overwrite_output()
                .run(quiet=True)
            )
        else:
            # No audio, just copy the video
            (
                ffmpeg
                .input(temp_output)
                .output(output_path, vcodec='copy')
                .overwrite_output()
                .run(quiet=True)
            )
    
    # Verify the result
    final_frames = get_frame_count_ffmpeg_python(output_path)
    
    # Verify FPS is maintained
    final_probe = ffmpeg.probe(output_path, v='quiet')
    final_video_stream = next((stream for stream in final_probe['streams'] if stream['codec_type'] == 'video'), None)
    final_fps_str = final_video_stream['r_frame_rate']
    final_fps_num, final_fps_den = map(int, final_fps_str.split('/'))
    final_fps = final_fps_num / final_fps_den
    final_duration = float(final_video_stream['duration'])
    
    return {
        'original_frames': current_frames,
        'target_frames': target_frames,
        'final_frames': final_frames,
        'original_fps': original_fps,
        'final_fps': final_fps,
        'original_duration': original_duration,
        'final_duration': final_duration,
        'scale_factor': scale_factor,
        'method_used': 'precise_maintain_fps',
        'fps_maintained': abs(final_fps - original_fps) < 0.01,
        'frame_count_success': final_frames == target_frames
    }

def scale_video_frames_precise_simple(input_path, output_path, target_frames):
    """
    Simplified version that definitely maintains FPS by using frame duplication/selection.
    """
    if not os.path.exists(input_path):
        raise FileNotFoundError(f"Input video file not found: {input_path}")
    
    current_frames = get_frame_count_ffmpeg_python(input_path)
    scale_factor = target_frames / current_frames
    
    print(f"Original frames: {current_frames}")
    print(f"Target frames: {target_frames}")
    print(f"Scale factor: {scale_factor:.3f}")
    
    # Get original FPS
    probe = ffmpeg.probe(input_path, v='quiet')
    video_stream = next((stream for stream in probe['streams'] if stream['codec_type'] == 'video'), None)
    fps_str = video_stream['r_frame_rate']
    fps_num, fps_den = map(int, fps_str.split('/'))
    original_fps = fps_num / fps_den
    
    if scale_factor > 1:
        # Need to duplicate frames - use setpts to slow down playback
        # This maintains FPS but increases duration
        setpts_factor = 1 / scale_factor
        (
            ffmpeg
            .input(input_path)
            .video
            .filter('setpts', f'{setpts_factor}*PTS')
            .output(output_path, vcodec='libx264', acodec='copy', r=original_fps)
            .overwrite_output()
            .run(quiet=True)
        )
    else:
        # Need to remove frames - select every nth frame
        select_interval = int(1 / scale_factor)
        (
            ffmpeg
            .input(input_path)
            .video
            .filter('select', f'not(mod(n,{select_interval}))')
            .filter('setpts', 'N/FRAME_RATE/TB')
            .output(output_path, vcodec='libx264', acodec='copy', r=original_fps)
            .overwrite_output()
            .run(quiet=True)
        )
    
    # Verify results
    final_frames = get_frame_count_ffmpeg_python(output_path)
    final_probe = ffmpeg.probe(output_path, v='quiet')
    final_video_stream = next((stream for stream in final_probe['streams'] if stream['codec_type'] == 'video'), None)
    final_fps_str = final_video_stream['r_frame_rate']
    final_fps_num, final_fps_den = map(int, final_fps_str.split('/'))
    final_fps = final_fps_num / final_fps_den
    
    return {
        'original_frames': current_frames,
        'target_frames': target_frames,
        'final_frames': final_frames,
        'original_fps': original_fps,
        'final_fps': final_fps,
        'scale_factor': scale_factor,
        'method_used': 'simple_maintain_fps',
        'fps_maintained': abs(final_fps - original_fps) < 0.01,
        'success': abs(final_frames - target_frames) <= 1
    }

# Example usage
if __name__ == "__main__":
    input_video = "input.mp4"  # Replace with your input file
    output_video = "output_scaled_maintain_fps.mp4"  # Output file
    target_frame_count = 10  # Desired number of frames
    
    try:
        # Method 1: Precise frame control with FPS maintenance
        result = scale_video_frames_precise(input_video, output_video, target_frame_count)
        print(f"Precise scaling result: {result}")
        print(f"FPS maintained: {result['fps_maintained']}")
        print(f"Original FPS: {result['original_fps']:.2f}, Final FPS: {result['final_fps']:.2f}")
        
        # Method 2: Simpler approach (alternative)
        # result = scale_video_frames_precise_simple(input_video, output_video, target_frame_count)
        # print(f"Simple scaling result: {result}")
        
    except Exception as e:
        print(f"Error: {e}")

for f in glob.glob("/Users/l_y_o/Work/godmodeai_video_training25/training_data/sfvskof/data/training_videos/shoryuken/*.mp4"):
    video_file = f
    frames = scale_video_frames_precise(video_file, f.replace("shoryuken", "shoryuken_scaled"), 23)
    #frames = scale_video_frames(video_file, f.replace(".mp4", "_scaled.mp4"), 23, method='duplicate')
    #print(f"Number of frames: {f} - {frames}")
```
