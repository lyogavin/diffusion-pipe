```python
target_path = '/Users/l_y_o/Work/godmodeai_video_training25/training_data/sfvskof/data/chars/training_videos'
```

```python
import glob
from tqdm import tqdm
import os, json

all_animations = set()

for chdir in tqdm(chars):
    json_file = chdir.replace(".txt", "_animations.json")
    if not os.path.exists(json_file):
        print("not found")
        continue
    with open(json_file, 'r') as jf:
        videos = json.load(jf)
    for l in videos:
        all_animations.add(l['animation'])
```

```python
import shutil
from pathlib import Path
animation_to_info = {}

for chdir in tqdm(chars):
    json_file = chdir.replace(".txt", "_animations.json")
    if not os.path.exists(json_file):
        print("not found")
        continue
    with open(json_file, 'r') as jf:
        videos = json.load(jf)
    for l in videos:
        #all_animations.add(l['animation'])
        source_file = f"{os.path.dirname(chdir)}/{l['video']}"
        p = Path(chdir)
        char_name = p.name.split(".")[0]
        
        # Generate a version 4 UUID (random)
        random_uuid = uuid.uuid4()
        random_uuid = str(random_uuid)[-8:]

        shutil.copy(source_file, f"{target_path}/{l['animation']}/{char_name}_{random_uuid}.mp4")
        
        if l['animation'] not in animation_to_info:
            animation_to_info[l['animation']] = []
        ll = l.copy()
        ll['video'] = f"{char_name}_{random_uuid}.mp4"
        animation_to_info[l['animation']].append(ll)
```

```python
def get_video_prompt(video_path, animation):
    print(f"uploading {video_path}, {os.path.isfile(video_path)}, {os.fspath(video_path)}")
    myfile = client.files.upload(file=video_path)
    time.sleep(2)


    prompt = f"Given this gaming {animation} animation of a character in side view with solid color background, we want to generate a video generation prompt that can be used to generate this video. In the prompt, describe the character in detail and the action the character is doing. Don't go beyond 230 words. Only output the prompt, nothing else."

    RETRY = 5
    for i in range(RETRY):
        try:
            response = client.models.generate_content(
                model="gemini-2.5-pro-preview-05-06", 
                contents=[myfile, prompt]
            )
            break
        except Exception as e:
            print(f"error: {e}")
            time.sleep(2)
    return response.text
                
```
